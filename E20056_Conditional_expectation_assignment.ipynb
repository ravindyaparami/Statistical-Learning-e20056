The Law of Total Variance shows that the "Optimal Error" is always less than the "Global Error."

# Q2) Conditional Density of the Multivariate Gaussian

Let $\mathbf{Z}$ be the concatenated random matrix
$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$.


The random variable $\mathbf{Z}$ is said to have a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

For $\mathbf{Z}$ to possess a **joint density function** (meaning its distribution is absolutely continuous with respect to the Lebesgue measure on $\mathbb{R}^{n+m}$), its covariance matrix must be non-singular (positive definite).


The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given by:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right).$$

Here:

* **Mean ($\boldsymbol{\mu}$):** $\boldsymbol{\mu} = E[\mathbf{Z}] = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix} \in \mathbb{R}^{n+m}$, where $\boldsymbol{\mu}_X \in \mathbb{R}^n$ and $\boldsymbol{\mu}_Y \in \mathbb{R}^m$.
* **Covariance Matrix ($\boldsymbol{\Sigma}$):** A symmetric, positive definite matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{(n+m) \times (n+m)}$ partitioned as:
$$\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$
    * $\boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X})$ (size $n \times n$)
    * $\boldsymbol{\Sigma}_{YY} = \text{Var}(\mathbf{Y})$ (size $m \times m$)
    * $\boldsymbol{\Sigma}_{XY} = \text{Cov}(\mathbf{X}, \mathbf{Y}) = \boldsymbol{\Sigma}_{YX}^T$ (size $n \times m$)
* **Determinant ($|\boldsymbol{\Sigma}|$):** The normalization factor ensuring the total integral is 1.

1. Show that the conditional distribution is given by $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right),$$
where
* $\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$

Answer:

Let$$Z = \begin{bmatrix} X \\ Y \end{bmatrix} \in \mathbb{R}^{n+m}$$be jointly Gaussian:$$Z \sim \mathcal{N}(\mu, \Sigma)$$with the partitioned mean vector and covariance matrix:$$\mu = \begin{bmatrix} \mu_X \\ \mu_Y \end{bmatrix}, \qquad \Sigma = \begin{bmatrix} \Sigma_{XX} & \Sigma_{XY} \\ \Sigma_{YX} & \Sigma_{YY} \end{bmatrix}$$where $\Sigma$ is positive definite. We want to derive the conditional density $f_{X|Y}(x|y)$.Step 1: Start from the definition of conditional density$$f_{X|Y}(x|y) = \frac{f_{X,Y}(x,y)}{f_Y(y)}$$The joint density $f_{X,Y}(x,y)$ is given by:$$f_{X,Y}(x,y) = \frac{1}{(2\pi)^{(n+m)/2}|\Sigma|^{1/2}} \exp\left( -\frac12 \begin{bmatrix} x-\mu_X \\ y-\mu_Y \end{bmatrix}^{T} \Sigma^{-1} \begin{bmatrix} x-\mu_X \\ y-\mu_Y \end{bmatrix} \right)$$Step 2: Marginal distribution of $Y$Since marginals of a multivariate Gaussian distribution are also Gaussian:$$Y \sim \mathcal{N}(\mu_Y, \Sigma_{YY})$$Thus, its marginal density is:$$f_Y(y) = \frac{1}{(2\pi)^{m/2}|\Sigma_{YY}|^{1/2}} \exp\left( -\frac12 (y-\mu_Y)^T \Sigma_{YY}^{-1} (y-\mu_Y) \right)$$Step 3: Form the ratioSubstituting the joint and marginal densities into the conditional density formula:$$f_{X|Y}(x|y) = \frac{ \frac{1}{(2\pi)^{(n+m)/2}|\Sigma|^{1/2}} e^{-\frac12 Q(x,y)} }{ \frac{1}{(2\pi)^{m/2}|\Sigma_{YY}|^{1/2}} e^{-\frac12 R(y)} }$$where the quadratic forms are defined as:$$Q(x,y) = \begin{bmatrix} x-\mu_X \\ y-\mu_Y \end{bmatrix}^{T} \Sigma^{-1} \begin{bmatrix} x-\mu_X \\ y-\mu_Y \end{bmatrix}$$$$R(y) = (y-\mu_Y)^T\Sigma_{YY}^{-1}(y-\mu_Y)$$Simplifying the constant terms yields:$$f_{X|Y}(x|y) = \frac{|\Sigma_{YY}|^{1/2}}{(2\pi)^{n/2}|\Sigma|^{1/2}} \exp\left( -\frac12(Q(x,y)-R(y)) \right)$$Step 4: Complete the quadratic formUsing block matrix inversion properties (or by completing the square), the difference in the exponent simplifies to a quadratic form purely in terms of $x$:$$Q(x,y) - R(y) = (x-\mu^*)^T \Sigma_{X|Y}^{-1} (x-\mu^*)$$where:$$\mu^* = \mu_X + \Sigma_{XY}\Sigma_{YY}^{-1}(y-\mu_Y)$$$$\Sigma_{X|Y} = \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}$$The matrix $\Sigma_{X|Y}$ is known as the Schur complement of $\Sigma_{YY}$ in $\Sigma$.Step 5: Simplify the determinant termUsing the block determinant identity:$$|\Sigma| = |\Sigma_{YY}| \cdot |\Sigma_{X|Y}|$$Taking the square root and rearranging gives:$$\frac{|\Sigma_{YY}|^{1/2}}{|\Sigma|^{1/2}} = \frac{1}{|\Sigma_{X|Y}|^{1/2}}$$Final ResultCombining the simplified constants and exponent yields the standard multivariate Gaussian density function:$$f_{X|Y}(x|y) = \frac{1}{(2\pi)^{n/2}|\Sigma_{X|Y}|^{1/2}} \exp\left( -\frac12 (x-\mu^*)^T \Sigma_{X|Y}^{-1} (x-\mu^*) \right)$$where:$$\mu^* = \mu_X + \Sigma_{XY}\Sigma_{YY}^{-1}(y-\mu_Y)$$$$\Sigma_{X|Y} = \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}$$Hence, the conditional distribution is:$$X \mid Y=y \sim \mathcal{N}\left(\mu_X + \Sigma_{XY}\Sigma_{YY}^{-1}(y-\mu_Y), \; \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}\right)$$

# Q3) Continuous Fault Monitoring of a Spindle Motor

A factory monitors the health of a high-speed CNC spindle. The critical variable is the **Internal Temperature ($X$)**, but because the motor is sealed, they can only observe the **Total Spindle Power ($Y$)**.

**The Setup:**
1.  **Hidden Temperature ($X$):** The standardized temperature deviation follows a standard normal distribution:
    $$X \sim \mathscr{N}(0, 1)$$
2.  **Mechanical Load ($Q$):** The torque required to cut material creates "noise" in the power reading. It is independent of temperature and follows:
    $$Q \sim \mathscr{N}(0, 4)$$
3.  **The Measurement ($Y$):** The power sensor $Y$ measures the combined effect of thermal load and mechanical torque:
$$Y = X + Q$$

---

**Answer the followings:**

**1. Joint Distribution**
Identify the distribution of the random varaible $(X, Y)$.
* Show that $Y \sim \mathscr{N}(0, 5)$.
* Find the covariance $\text{Cov}(X, Y)$.

**2. Information Geometry**
Explain why $X$ is **not** $\sigma(Y)$-measurable. In practical terms, if the power meter shows $Y = 5$, why can't the engineer know for certain if the motor is overheating ($X$ is high) or if the material being cut is simply very hard ($Q$ is high)?

**3. Conditional Expectation (The Best Estimate)**
Compute the conditional expectation $E[X \mid Y]$.
* Using the property for joint normals: $E[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y)$.
* Express this as a "signal-to-total-variance" ratio.

**4. Verification of the Residual**
Let $\widehat{X} = E[X \mid Y]$. Define the estimation error (residual) as $Z = X - \widehat{X}$.
* Prove that $E[Z] = 0$.
* Verify that the error $Z$ is uncorrelated with the observation $Y$ (the Orthogonality Principle).

**5. Information Theory: Differential Entropy**
* Compute the differential entropy $h(X)$.
* Compute the conditional differential entropy $h(X \mid Y)$.
* *Note: The variance of the conditional distribution $(X \mid Y=y)$ is given by $\sigma_{X|Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)}$.*

**6. Mutual Information**
Calculate $I(X; Y) = h(X) - h(X \mid Y)$.
* What percentage of the "uncertainty" about the temperature is removed by observing the power meter?
* How would this change if the mechanical noise $Q$ had a variance of $100$ instead of $4$?

**7. Diagnostic Interpretation**
The factory manager says: *"Since $Y = X + Q$, our best estimate of temperature should just be $Y$ itself."*
Using your result from Question 3, explain why $Y$ is a **biased** estimator of $X$ and why $E[X \mid Y]$ "shrinks" the measurement toward the mean (zero).



Answer:

**1. Joint Distribution of $(X, Y)$:**

Since $X$ and $Q$ are independent normal random variables, their sum $Y = X + Q$ is also normally distributed.
* **Mean:** $E[Y] = E[X] + E[Q] = 0 + 0 = 0$.
* **Variance:** Since $X \perp Q$, $\text{Var}(Y) = \text{Var}(X) + \text{Var}(Q) = 1 + 4 = 5$.
* **Covariance:** $\text{Cov}(X, Y) = E[XY] - E[X]E[Y] = E[X(X+Q)] = E[X^2] + E[XQ]$.
    * Since $E[X^2] = \text{Var}(X) = 1$ and $E[XQ] = E[X]E[Q] = 0$:
    * **$\text{Cov}(X, Y) = 1$**.

The joint distribution is:
$$\begin{pmatrix} X \\ Y \end{pmatrix} \sim \mathscr{N} \left( \begin{pmatrix} 0 \\ 0 \end{pmatrix}, \begin{pmatrix} 1 & 1 \\ 1 & 5 \end{pmatrix} \right)$$

**2. Information Geometry:**

There is no deterministic function $g$ such that $X = g(Y)$. For any observed $Y=y$, there are infinitely many pairs $(x, q)$ such that $x+q=y$. So, X is no $σ$(Y) measurable.
A high power reading ($Y$) is ambiguous.(could be a "Thermal Fault-high $X$ or simply a "Hard Material" - high $Q$). The sensor $Y$ cannot distinguish between these two physical states.

**3. Conditional Expectation **

$$E[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y) = 0 + \frac{1}{5}(y - 0)$$
$$\mathbf{E[X \mid Y] = 0.2 Y}$$

Because the noise ($Q$) is 4 times more volatile than the signal ($X$), we only "trust" 20% of the power fluctuation as being caused by temperature.


**4. Verification of the Residual:**

Let $\widehat{X} = 0.2 Y$. The residual is $Z = X - 0.2 Y$.
* **Mean:** $E[Z] = E[X] - 0.2 E[Y] = 0 - 0 = 0$.
* **Orthogonality:** $\text{Cov}(Z, Y) = E[ZY] = E[(X - 0.2Y)Y] = E[XY] - 0.2 E[Y^2]$.
    * $E[XY] = 1$ and $E[Y^2] = 5$.
    * $\text{Cov}(Z, Y) = 1 - 0.2(5) = 1 - 1 = 0$.
The error is uncorrelated with our observation whicjh means that we have extracted all possible information from $Y$.

**5. Differential Entropy:**

Recall $h(\mathscr{N}(\mu, \sigma^2)) = \frac{1}{2}\ln(2\pi e \sigma^2)$.
* **Marginal $h(X)$:** With $\sigma_X^2 = 1$:
$$h(X) = \frac{1}{2}\ln(2\pi e) \approx 1.4189 \text{ nats}$$
* **Conditional $h(X|Y)$:** The conditional variance is
$$\sigma_{X|Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)} = 1 - \frac{1}{5} = 0.8.$$
$$h(X|Y) = \frac{1}{2}\ln(2\pi e \cdot 0.8) \approx 1.3073 \text{ nats}$$

---

**6. Mutual Information:**

$$I(X; Y) = h(X) - h(X|Y) = \frac{1}{2}\ln(1) - \frac{1}{2}\ln(0.8) = \frac{1}{2}\ln\left(\frac{1}{0.8}\right) = \frac{1}{2}\ln(1.25) \approx \mathbf{0.1116 \text{ nats}}$$

The variance of $1.0$ changed to $0.8$ That is a **20% reduction**. If the noise $Q$ increased to $\sigma_Q^2 = 100$, $I(X;Y)$ would approach 0, as the temperature signal would be "drowned out."

**7. Diagnostic Interpretation:**

* If we use $X = Y$, $E[(X-Y)^2] = E[Q^2] = 4$.
* By using $E[X|Y] = 0.2Y$, error variance is only $0.8$.

Why $Y$ is a Weak/Biased Estimator of $X$

The manager’s guess ($\hat{X} = Y$) treats all power spikes as pure temperature changes.While $Y$ matches $X$ on average across infinite runs ($\mathbb{E}[Y] = \mathbb{E}[X] = 0$), it is a highly inaccurate estimator for any specific moment. Because the torque noise ($Q$) is 4 times more volatile than the temperature signal ($X$), using $Y$ directly floods your estimate with noise, resulting in a massive error variance of 4.0.2.

Why $\mathbb{E}[X \mid Y]$ "Shrinks" Toward Zero

From Question 3, the optimal minimum mean-square estimate is:$$\mathbb{E}[X \mid Y] = 0.2Y$$The $0.2$ multiplier is a shrinkage factor that scales the raw reading back toward the baseline average (zero) because:The 20% Trust Rule: Since temperature accounts for only $\frac{1}{1+4} = 20\%$ of the total variance in $Y$, we mathematically only "trust" $20\%$ of any power fluctuation as being caused by temperature.Noise vs. Signal: If power spikes violently, it is statistically far more likely that the machine simply hit a hard piece of material ($Q$) rather than the tool suddenly melting ($X$).🛠️ Operational Impact: Shrinkage prevents False Alarms. It stops the automated factory system from triggering emergency cooling shutdowns every time the machine experiences routine mechanical resistance. This brings the error variance down from 4.0 to just 0.8.
